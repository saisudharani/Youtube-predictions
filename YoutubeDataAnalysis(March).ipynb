```python
c = widgets.Dropdown(
    options=video['Genre'].unique(),
    value='Music',
    description='Genre:',
    disabled=False,
)

def graph(a):
  dr = video[video['Genre']==a]
  gt = dr.groupby('channelTitle')
  gt = gt.sum().sort_values(by = 'view_count',ascending = False).head()
  sns.barplot(y = gt.index,x = gt['view_count'])
  plt.title(a)
  plt.xlabel('Views')
  plt.ylabel('Channel Name')
  plt.show()
```

```python
widgets.interact(graph,a=c)
```

```python
g = widgets.Dropdown(
    options=video['channelTitle'].unique(),
    value='NBA',
    description='Channel:',
    disabled=False,
)

def graph2(z):
  ht = video[video['channelTitle']==z]
  ht = ht.sort_values(by = 'view_count',ascending = False).head()
  sns.barplot(y = 'title',x = 'view_count', data = ht)
  plt.title(z)
  plt.xlabel('Views')
  plt.ylabel('Video Name')
  plt.show()
```

```python
widgets.interact(graph2,z=g)
```

```python
k1 = widgets.Dropdown(
    options=video['Genre'].unique(),
    value='Music',
    description='Genre',
    disabled=False,
)

def genre(m):
  gen = video[video['Genre']==m]
  k2 = widgets.Dropdown(
    options=gen['channelTitle'].unique(),
    value=gen['channelTitle'].unique()[0],
    description='channel',
    disabled=False,
  )
  def channel(d):
    ht = gen.sort_values(by = 'view_count',ascending = False).head()
    sns.barplot(y = 'title',x = 'view_count', data = ht)
    plt.title(d)
    plt.xlabel('Views')
    plt.ylabel('Video Name')
    plt.show()

  widgets.interact(channel,d = k2)
```

```python
widgets.interact(genre,m = k1)
```

```python
c = widgets.Dropdown(
    options=video['Genre'].unique(),
    value=video['Genre'].unique()[0],
    description='Genre:',
    disabled=False,
)

c1 = widgets.IntSlider(
    value=7,
    min=0,
    max=10000000,
    step=1,
    description='Views:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)
c2 = widgets.IntSlider(
    value=7,
    min=0,
    max=100000,
    step=1,
    description='comment:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)
c3 = widgets.IntSlider(
    value=7,
    min=0,
    max=1000000,
    step=1,
    description='dislikes:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

def predictedval(a,a1,a2,a3):
  print(a,a1,a2,a3)
  id = video[video['Genre']==a]['categoryId'].iloc[0]
  print('Predicted Likes count:',int(model.predict([[id,a1,a3,a2]])[0]))

```
